# Lists

lists: a type of data structure that come hold multiple pieces of data
- lists are defined by square brackets []
- similar to arrays in other languages, except lists can hold multiple data types and are dynamic in size
- lists are mutable, meaning that they can be changed after they are created, and can be modified, deleted, or change elememts

creating a list

nameOfList = []

```python
# list of test scores for a class test

testScores = [100, 90, 80, 70, 60]
#if we need to access or chagne a single item/element in a list, we use the index of the element to grab it.
#lists start at the 0 index posistion (index means where the elements are located in the list by its order) 100 is 0, 90 is 1, 80 is 2 etc...

print(testScores[0]) #will print the 100 from our list
print(testScores[1]) #will print the 90

#can also change elements
testScores[0] = 95 # will change the 100 to 95

#we want to use lists to group things together as well as be able to iterate over them to change or access the elements in thet list (iterate 
# ie use a loop to go through each element in the list)

#if we need to display or modify items in a list, we use a for loop

#in python there are a few types of for loops

#for in loop - for items in this list do the following thing
#we are going to use a for loop to display all of the test scores in the list. To do so we create a variable place holder to represent each INDIVIDUAL
# score in the testScores lists

#for single item in list
for score in testScores:
    print(score)

#instead of printing the score, we can add to each score
for score in testScores:
    score += 5
    print(score)

#adding data to a list - we can add data to the list by using the append() method
newScore = 50
testScores.append(newScore) #will add it to the end by default
print(testScores)



#lets go through all of the list functions and methods that we can use with lists


#extend() - combines data from list to another
testScores2 = [10, 20, 90]
testScores.extend(testScores2) #will add at the end by default
print(testScores)

#removing from a list you can use the remove function/method
testScores.remove(10) #removes the first instance of the number 10 in the list.... Use the element and not the index
print(testScores)

#finding something in a list - this will give you a true or false value. then you can use if statements to remove/modify values
lookingFor = 80
#the "in" keyword is used to check if a value is IN a list
found = lookingFor in testScores
if found == True:
    print("Found the Value")
    testScores.remove(lookingFor)
    print(testScores)
else:
    print("Value not found")


#count() - counts the number of times a value appears in a list
print(testScores.count(90))

numbers = [10, 3, 7, 5, 9, 100]
#sort() - sorts the list in ascending order (smallest to largest)
numbers.sort()
print(numbers)

#sort in descending order
numbers.sort(reverse=True)
print(numbers)

#find the length of the list by using len() function
print(len(numbers))

#this is an alternative to a for in loop, in is similar to what you would see with a traditional for loop in other languages
for i in range(len(testScores)):
    print(i, "-", testScores[i]) #i represents the variable but also the index of the list, the first iteration i will be 0,  second is i, etc...
```

# Tuples
- are immutable (does not change) list of data... aka list of constants. once data is entered, it does not change
- tuples are faster than lists (use it when you can because its much faster)
- constant is a variable that does not change. delcared and used
- tuples are used for fixed data sets (Data that does not change) (ex. list of states)
- tuples have to be replaced to be changed

```python
# Here is an example of something that is imutable (does not change) - Strings

# When changing a string value, it is actually creating a new string

# Strings are technically arrays or lists of characters that represent an invidual character ot symbol

name = "Cris"

print(name)

# name[0] = "T" # this will not work because strings are immutable
name = "Tris" # this will work because it is creating a new string #not modifying but replacing
print(name)
```

```python
# tuples are created using parenthese - tuplename = (data1, data2, data3)
# tuples in general can be accessed the same way a lsit can be
states = ("IL", "KY", "OH")

#print tuples just like lists
for state in states:
    print(state)

# there is one type of modifying you can do you can combine tuples because it makes or overwrites a new tuple
states2 = ("FL", "GA", "AL")
states = states + states2
print(states)

```

# Dictionaries
- create key-value pairs that associate a characteristic/property with a value. Basically in python it is a hash table or an object
- similar to a class without methods ie an object literal
- an 'object' is anything with key-value pairs or properties
- dictionaries are mutable

- objects/dictionaries can be almost anything but in programming it usually refers to a wrapper that encapsulates related data together to describe 
some like in the following:
- object called Person - name, age
- object called Car - make, model, year

- dictionaries can group something by a list of items and associated values (like an actual dictionary)

```python
# lets look at the example below: there is a list of bank customers, and their bank balances
# dictionaries is created by using curly braces - dictionaryName = {key1:value1, key2:value2, key3:value3}
bankCustomers = {
    "Alice": 100,
    "Bob": 200,
    "Charlie": 300
}

print(bankCustomers["Bob"]) # will print 200 since Bob is the key loop (aka acts as an index)

# dictionaries can be accessed by the key, to display everything that we can use a loop
for customer in bankCustomers:
    print(customer, bankCustomers[customer]) #first iteration is Alice, 100. customer is the key, bankCustomer[customer] is the value where customer
    # acts as as an index (but called key lookup since its for dictionaries)

# can use dictionary like an object
personCris = {
    "name": "Cris",
    "age": 27,
    "city": "Fort Wayne"
}
for detail in personCris: #detail becomes the key. so first would be name, second is age, third is city
    print(detail, "-",  personCris[detail])

# ditionaries can be modified by adding a new key-value pair. if key is existing then the value will be replaced by the new one
# referring to an item by a key name and assign a value. if there is not a key name that matches then a new one will be added
# syntax -> dictionaryname[key] = value
bankCustomers["Bob"] = 400 #modifies the "Bob" value
print(bankCustomers)
bankCustomers["Cris"] = 1000 #adds the Cris key-value pair to our dictionary
print(bankCustomers)

# grab a single item from a list... call by a key name
# use .get method to grab data from a dictionary
print(bankCustomers.get("Alice")) #will give you the value which is 100

# delete item by key
# syntax - del dictionaryname[key]
# del is not a function, it is a keyword
del bankCustomers["Alice"] # will delete the key and value for alice
print(bankCustomers)

# finding items in a dictionary
# syntax -> key in dictionaryname
key = "Bob"
if key in bankCustomers:
    print(key, "is in the bankCustomers dictionary")

# matching name and values to find in a dictionary
findName = "Cris"
value = 1000
# for key value in dictionaryname.items()
# .items() -> casts each key value into individual items
for name, balance in bankCustomers.items(): #name, balance == key, value in the dictionary. It will always be key, value when dictioanryies have two variables in the a for loop
    if name == findName and balance == value:
        print(findName, "has a balance of", value)

print(bankCustomers.items())

for name, balance in bankCustomers.items(): # name, balance == key, value
    print(name, balance)

for name in bankCustomers: #name == key
    print(name)

for balance in bankCustomers.values(): # balance == value
    print(balance)
```
