# Class vs Instance Variable
- classes can represent real world objects with data and behavior that can be modified and changed
- if we want the attribute/data to be blank until we create a class then we can use instance variableds
- instance varible-> can be set or changed upon the creation of an object/instance
- class variable-> do not change upon the creation/instantiation of an object BUT can be changed afterwards if needed

```python
# create a class callled dog. Dog will have a couple of attributes - name and animal. Dog will always have the class variable of animal = "dog", but every dog can have a different name. so we use our instance variable (name) to set the name of the dog upon creation

class Dog:
    #class variable
    animal = "Dog"
    #anytime we create a new dog, it will always have animal = "Dog". If we need to change it later, we can

    #instance variables are used in arguments/parameters when creating an instance in order to set attributes upon creation
    #contructors: initalize the instance variables with some data that the user or other parts of the program provide upon creation. It is null until we give it data
    # constructors are special methods/functions
    # the self keyword is used when creating an instance variable
    #self = that individuals instance value or data (ie internal state)
    #__init__ = constructor keyword in python #init letting us know that the following are instanced variables

    #make a method or function in a class we just use the same syntax we do with a function
    def __init__(self, name, age): # age was added last to show example#self -> its own internal state #after self then we add a comma and put variables and more commas
        #name is going to be whatever you put in the parentheses
        self.name = name
        self.age = age #adding this one second to show how this works

    # common behavior can be created with functions
    def bark(self):
        print("Woof")

        

dog1 = Dog("Cris", 8)
print(dog1.name)
print(dog1.age)
#method call or object function call
dog1.bark()

dog2 = Dog("Spot", 10)
print(dog2.name)
print(dog2.age)
dog2.bark()

#there are class variables and they should be the same
print(dog1.animal)
print(dog2.animal)

#class variables can change afterwards after creating the object
dog2.animal = "Cat"
print(dog2.animal)

#use inputs to set internal state or instance variables
dog_name = input("What is the dogs name: ")
dog_age = int(input("What is the dogs age: "))

dog3 = Dog(dog_name, dog_age)
print(dog3.name)
print(dog3.age)
dog3.bark()



```

# Access Modifiers

- encapsulation --> wrap data and give them the ability to be accessible or inaccessible
- we can set variables so that other classes and outsdie data cannot interfere or change data unless specified or used with a method
- useful to limiting bugs
- we can limit access using functions so only certain groups/people/classes can manipulate data
- provide better security and robustness

```python
class Pet:
    def __init__(self, name, animal, age):
        #to make fields/attributes/variables private we use a double underscore aka a 'dunder' (__)
        self.__name = name
        self.__animal = animal
        self.__age = age

    #methods called getters/setters
    #getter retreives data
    #setters change/modify data

    #getters
    def get_name(self):
        return self.__name
    def get_animal(self):
        return self.__animal
    def get_age(self):
        return self.__age
    
    #setters
    def set_name(self, name):
        self.__name = name
    def set_animal(self, animal):
        self.__animal = animal
    def set_age(self, age):
        self.__age = age

pet1 = Pet("Spot", "dog", 5)
#run our getters
print(pet1.get_name())
print(pet1.get_animal())
print(pet1.get_age())

#change pets name, age, and type
pet1.set_name("Jack")
pet1.set_animal("Cat")
pet1.set_age(2)

#display new values
print(pet1.get_name())
print(pet1.get_animal())
print(pet1.get_age())
```

# inhertance and polymorphism prof lecture
- non-inheritance polymorphism
- polymorphism - functinos have the same name but do different things
- group by behavior

```python
#non-inheritance poly
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def move(self):
        print("VROOOOM")

class Boat:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def move(self):
        print("SAIL")

class Plane:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def move(self):
        print("FLY")

car1 = Car("Ford", "Mutang")
boat1 = Boat("Ibiza", "Touring 20")
plane1 = Plane("Boeing", "747")

car1.move() #this is polymorpheism because all of the functions have the same name but because its in a different function it means that they do different things
boat1.move()
plane1.move()
```

```python
# inhertiance based polymorphism -> Inherit a class and overrides functions or methods

#super or parent class
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    #default behavior
    def move(self):
        print("MOVE")

#inheritance and override the move method/function
class Car(Vehicle): #we're inheriting the Vehicle class so we dont have to repeat it in these classes
    def move(self):
        print("Vroom")

class Plane(Vehicle):
    def move(self): #this functions are getting overridden since our parent class has "MOVE" action in it
        print("FLY")

class Boat(Vehicle):
    def move(self):
        print("SAIL")


car1 = Car("Ford", "Mustang")
plane1 = Plane("Boeing", "747")
boat1 = Boat("Princess", "40M")

car1.move()
plane1.move()
boat1.move()
```

# decorators and methods as functions
- in python, a decorator is a function that takes another function as input and extends or modifies its behavior without explicitly modifying the original functions code
- decorators are a powerful way to add functionality to existing functions or classes, promoting code reusablility and seperation of concerns

- a decorator is a "wrapper" for a function that adds code to the function in order to make it do something extra

```python
def my_decorator(func):
    def wrapper():
        print("Something added before the function")
        func()
        print("Something added after the function")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()
```
